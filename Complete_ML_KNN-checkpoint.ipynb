#  step 1: import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# step 2: Read the dataset

```python
sns.get_dataset_names()
```

```python
dataset=sns.load_dataset('iris')
```

```python
dataset.head()
```

# step 3: identify the probelm 
## whether it is a classification problem/ Regression problem

```python
dataset['species'].unique()
```

# step 4: Data Preprocessing

```python
# NUll values
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
# object => in pandas = = > string/categorical features/boolean/Text
```

```python
# convert object types to numerical form ==> either float or integer
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
dataset['species']=encoder.fit_transform(dataset['species'])
```

```python
dataset.info()
```

```python
dataset.describe().plot()
```

# no outliers

```python
dataset.plot()
```

# separting dependent and independent features

```python
#indpendent features
x=dataset.drop(['species'],axis=1)
```

```python
x
```

```python
#dependent feature
y=dataset['species']
```

```python
y
```

```python
y.unique()
#(['setosa', 'versicolor', 'virginica']
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(x)
```

```python
x
```

```python
pd.DataFrame(x).plot()
```

# step 4: Deciding the model

```python
sns.pairplot(dataset,hue='species')
```

# step 5: Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Step 6: Buliding the model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=3)
```

```python
# training the model
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)* 100
```

# 97% accuracy for Training our model

# step 7: model predictions

```python
y_pred=model.predict(x_test)
```

```python
y_pred.reshape(-1,1)
```

```python
y_test
```

# step 8: evaluation of our model

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)* 100
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# 16+17+11 ==> correct predictions
# 0+0+0+0+0+1 ==> incorrect predictions
```
